```python
import os
import pandas as pd 
from shutil import copyfile
```

```python
labels = pd.read_csv("dataset/label.csv")
```

```python
labels
```

```python
file_names_list = list(labels["file_name"])
tumor_type_list = list(labels["label"])

no_tumor_list = []
tumor_list = []

no_tumor_count = 0
tumor_count = 0

for i in range(len(labels)):
    if tumor_type_list[i] == "no_tumor":
        no_tumor_count = no_tumor_count+1 # just for checking none are missed
        no_tumor_list.append(file_names_list[i])
    else:
        tumor_count = tumor_count+1 # just for checking none are missed
        tumor_list.append(file_names_list[i])
        
i=0
while i<no_tumor_count:
    copyfile("dataset/image/"+no_tumor_list[i], "dataset/split_data/binary/no_tumor/"+no_tumor_list[i])
    i=i+1

i=0
while i<tumor_count:
    copyfile("dataset/image/"+tumor_list[i], "dataset/split_data/binary/tumor/"+tumor_list[i])
    i=i+1
```

```python
file_names_list = list(labels["file_name"])
tumor_type_list = list(labels["label"])

no_tumor_list = []
meningioma_tumor_list = []
glioma_tumor_list = []
pitulitary_tumor_list = []

no_tumor_count = 0
meningioma_count = 0
glioma_count = 0
pitulitary_count = 0

for i in range(len(labels)):
    if tumor_type_list[i] == "no_tumor":
        no_tumor_count = no_tumor_count+1 # just for checking none are missed
        no_tumor_list.append(file_names_list[i])
    
    if tumor_type_list[i] == "meningioma_tumor":
        meningioma_count = meningioma_count+1 # just for checking none are missed
        meningioma_tumor_list.append(file_names_list[i])
    
    if tumor_type_list[i] == "glioma_tumor":
        glioma_count = glioma_count+1 # just for checking none are missed
        glioma_tumor_list.append(file_names_list[i])
    
    if tumor_type_list[i] == "pituitary_tumor":
        pitulitary_count = pitulitary_count+1 # just for checking none are missed
        pitulitary_tumor_list.append(file_names_list[i])

print(no_tumor_count+meningioma_count+glioma_count+pitulitary_count)
i=0
print("starting no tumor list, count is "+ str(no_tumor_count))
while i<no_tumor_count:
    print("moving " +no_tumor_list[i] + " into no tumor")
    copyfile("dataset/image/"+no_tumor_list[i], "dataset/split_data/multi-class/no_tumor/"+no_tumor_list[i])
    i=i+1

i=0
print("starting meningioma list, count is "+ str(meningioma_count)+ " -- " +str(i))
while i<meningioma_count:
    print("moving " +meningioma_tumor_list[i] + " into meningioma tumor")
    copyfile("dataset/image/"+meningioma_tumor_list[i], "dataset/split_data/multi-class/meningioma_tumor/"+meningioma_tumor_list[i])
    i=i+1
    

i=0
print("starting glioma list, count is "+ str(glioma_count))
while i<glioma_count:
    print("moving " +glioma_tumor_list[i] + " into glioma tumor")
    copyfile("dataset/image/"+glioma_tumor_list[i], "dataset/split_data/multi-class/glioma_tumor/"+glioma_tumor_list[i])
    i=i+1
    

i=0
print("starting pitulitary_count list, count is "+ str(pitulitary_count))
while i<pitulitary_count:
    print("moving " +pitulitary_tumor_list[i] + " into pitulitary tumor")
    copyfile("dataset/image/"+pitulitary_tumor_list[i], "dataset/split_data/multi-class/pitulitary_tumor/"+pitulitary_tumor_list[i])
    i=i+1
```
