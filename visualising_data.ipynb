```python
def plotfolder(folder, num_to_plot, save_dir_and_name, title, random_images):
    for i, file in enumerate(os.listdir(folder)[0:num_to_plot]):
        
        if not random_images:
            fullpath = folder + file
        else:
            fullpath = folder + os.listdir(folder)[random.randint(0,len(os.listdir(folder)))]
            
        img=mpimg.imread(fullpath)
        plt.subplot(int(math.sqrt(num_to_plot)),int(math.sqrt(num_to_plot)), i+1)
        
        plt.tick_params(
        axis='both',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        labelbottom=False,  # labels along the bottom edge are off
        left=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelleft=False) # labels along the bottom edge are off
        
        plt.suptitle(title)     
        plt.imshow(img)    
        plt.savefig(save_dir_and_name)
```

```python
plotfolder("dataset/split_data/multi-class/pitulitary_tumor/", 4, "dataset/pitulitary_tumor4.png", "pitulitary tumor", False)
```
